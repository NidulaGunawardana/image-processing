### Computing Histograms

```python
import cv2 as cv
import numpy as np


def rescaleFrame(frame,scale=0.75):
    """
    This method will work for any kind of video photo or live video
    """
    width = int(frame.shape[1]*scale)
    height = int(frame.shape[0]*scale)
    
    dimentions = (width,height)
    
    return cv.resize(frame,dimentions,interpolation=cv.INTER_AREA)

img = cv.imread('../1. Read image and video/images/image1.jpg')
scaled_img = rescaleFrame(img,0.1)
cv.imshow('Color image',scaled_img)


cv.waitKey(0)
```

```python
blank = np.zeros(scaled_img.shape[:2], dtype="uint8")
cv.imshow("Blank", blank)

cv.waitKey(0)
```

```python
# gray = cv.cvtColor(scaled_img,cv.COLOR_BGR2GRAY)
# cv.imshow("Gray image",gray)

# cv.waitKey(0)
```

```python
circle = cv.circle(
    blank, (scaled_img.shape[1] // 2 , scaled_img.shape[0] // 2),100, 255, -1
)

mask = cv.bitwise_and(scaled_img,scaled_img,mask=circle)

cv.imshow("Mask",mask)

cv.waitKey(0)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title("Color Histogram")
plt.xlabel('Bins')
plt.ylabel('No of pixels')




colors = ('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([scaled_img],[i],circle,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,256])

plt.show()
```
